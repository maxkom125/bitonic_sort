```python
import matplotlib.pyplot as plt
```

```python
c = 0.179
degree = 6
```

Формула, описанная в документе

```python
def theory(x0, degree):
    return x0 / (x0 / 2**degree + (2**(degree+1) - 2 - degree) * c)
```

```python
x = [20.217673,
18.453157,
10.670435,
5.432000,
2.882740,
2.160427,
1.761642]
```

```python
[x[0]/i for i in x]
```

```python
plt.plot([i for i in range(1, 2**(len(x)-1))], label='ideal')
plt.plot([2**i for i in range(len(x))], [x[0]/i for i in x],'go-', label='program')
plt.plot([2**i for i in range(len(x))], [theory(x[0], i) for i in range(len(x))],'yo-', label='theory')
plt.title('Acceleration (degree = 24)')
#plt.text(2, 0.65, r'$\cos(2 \pi t) \exp(-t)$', fontdict=font)
plt.xlabel('processes')
plt.ylabel('acceleration')
plt.legend()
plt.grid(True)
plt.show()
```
